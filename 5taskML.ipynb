```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

1. Загрузите обучающую и тестовую выборки из файлов perceptron-
train.csv и perceptron-test.csv. Целевая переменная записана в пер-вом столбце, признаки — во втором и третьем.

```python
data_train = pd.read_csv('perceptron-train.csv', header=None)
data_test = pd.read_csv('perceptron-test.csv', header=None)
X_train = data_train.iloc[:,1:]
y_train = data_train.iloc[:,0]
X_test = data_test.iloc[:,1:]
y_test = data_test.iloc[:,0]
print(X_train)
print(X_test)
```

2. Обучите персептрон со стандартными параметрамии random_state=241.

```python
clf = Perceptron(random_state=241)
clf.fit(X_train, y_train)
```

3. Подсчитайте качество (долю правильно классифицированных объ-
ектов, accuracy) полученного классификатора на тестовой выборке.

```python
predictions = clf.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(accuracy)
```

4. Нормализуйте обучающую и тестовую выборку с помощью класса
StandardScaler.

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

5. Обучите персептрон на новых выборках. Найдите долю правиль-
ных ответов на тестовой выборке.

```python
clf_scaled = Perceptron(random_state=241)
clf_scaled.fit(X_train_scaled, y_train)
predictions_scaled = clf_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, predictions_scaled)
print(accuracy_scaled)
```

6. Найдите разность между качеством на тестовой выборке после нор-
мализации и качеством до нее. Это число и будет ответом на зада-
ние.

```python
print(f'{round(accuracy_scaled - accuracy,3)} - разница качества точности классификаторов до и после нормализации')
```
